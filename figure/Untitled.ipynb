```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
a = np.linspace(-5,5,1000)
b = np.exp(- ((a-2) * (a-2)) ) + np.exp(- ((a+2) * (a+2)) )
plt.plot(a,b,label="Double Center")
c = np.exp(- 0.2 *((a) * (a)) )
plt.plot(a,c,label="Single Center")
plt.legend()
plt.savefig("centers_gaussian.eps")
```
